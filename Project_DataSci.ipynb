## Random Forest

```python
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y.values, test_size=0.01, random_state=42)

# Define the hyperparameters for the RandomForestRegressor
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30, 40],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1],
    'max_features': ['sqrt'],
    'bootstrap': [False]
}

# Create the RandomForestRegressor
regressor_rf = RandomForestRegressor(random_state=0)

# Perform GridSearchCV
gs = GridSearchCV(regressor_rf, param_grid, scoring='neg_mean_squared_error', cv=5)
gs.fit(X_train, y_train)
regressor_rf = gs.best_estimator_

regressor_rf.fit(X_train, y_train)
# Print the best hyperparameters
print("Best Hyperparameters:", gs.best_params_)

# Make predictions
y_train_pred = regressor_rf.predict(X_train)
y_test_pred = regressor_rf.predict(X_test)

# Calculate and print the root mean squared error for training set
error_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
print("Root Mean Squared Error (Training): {:,.02f}".format(error_train))

# Calculate and print the root mean squared error for testing set
error_test = np.sqrt(mean_squared_error(y_test, y_test_pred))
print("Root Mean Squared Error (Testing): {:,.02f}".format(error_test))

# Print the best number of estimators
best_n_estimators = gs.best_estimator_.n_estimators
print("Best Number of Estimators:", best_n_estimators)

# Calculate and print R^2 for training set
r2_train = r2_score(y_train, y_train_pred)
print("R^2 (Training): {:.2f}".format(r2_train))

# Calculate and print R^2 for testing set
r2_test = r2_score(y_test, y_test_pred)
print("R^2 (Testing): {:.2f}".format(r2_test))
```

## Gradient Boost

```python
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y.values, test_size=0.01, random_state=40)

# Define the hyperparameters for the GradientBoostingRegressor
param_grid = {
    'n_estimators': [500],
    'learning_rate': [0.01, 0.1, 0.2, 0.3],
    'max_depth': [3, 5, 7, 9],
    'min_samples_split': [2, 5, 10, 20],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2', None]
}
regressor_gb = GradientBoostingRegressor(random_state=0)

gs = GridSearchCV(regressor_gb, param_grid, scoring='neg_mean_squared_error', cv=5)
gs.fit(X_train, y_train)
regressor_gb = gs.best_estimator_

regressor_gb.fit(X_train, y_train)

print("Best Hyperparameters:", gs.best_params_)

y_train_pred = regressor_gb.predict(X_train)
y_test_pred = regressor_gb.predict(X_test)

error_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
print("Root Mean Squared Error (Training): {:,.02f}".format(error_train))

error_test = np.sqrt(mean_squared_error(y_test, y_test_pred))
print("Root Mean Squared Error (Testing): {:,.02f}".format(error_test))

best_n_estimators = gs.best_estimator_.n_estimators
print("Best Number of Estimators:", best_n_estimators)

r2_train = r2_score(y_train, y_train_pred)
print("R^2 (Training): {:.2f}".format(r2_train))

r2_test = r2_score(y_test, y_test_pred)
print("R^2 (Testing): {:.2f}".format(r2_test))
```

## Decision Tree

```python
# decision tree
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score

X_train, X_test, y_train, y_test = train_test_split(X, y.values, test_size=0.04, random_state=33)

regressor_dt = DecisionTreeRegressor(random_state=36)

param_grid = {
    'max_depth': [14, 15, 16, 17, 18, 19, 20],
    'min_samples_split': [2, 3, 4],
    'min_samples_leaf': [1, 2, 3],
    'max_features': [ 'sqrt', 'log2']
}

grid_search = GridSearchCV(regressor_dt, param_grid, scoring='neg_mean_squared_error', cv=7, n_jobs=-1)
grid_search.fit(X_train, y_train)
# Get the best estimator
best_regressor_dt = grid_search.best_estimator_

# Print best Hyperparameters
print("Best Hyperparameters:\n", grid_search.best_params_)

# Predict
y_pred = best_regressor_dt.predict(X_test)

# Calculate and print RMSE on training set & RMSE Testing
rmse_train = np.sqrt(mean_squared_error(y_train, best_regressor_dt.predict(X_train)))
print("Root Mean Squared Error (Training): {:.2f}".format(rmse_train))

rmse_test = np.sqrt(mean_squared_error(y_test, y_pred))
print("Root Mean Squared Error (Testing): {:.2f}".format(rmse_test))

# Calculate and print R^2 on training set
y_train_pred = best_regressor_dt.predict(X_train)
r2_train = r2_score(y_train, y_train_pred)
print("R^2 (Training): {:.2f}".format(r2_train))

# Calculate and print R^2 on testing set
r2_test = r2_score(y_test, y_pred)
print("R^2 (Testing): {:.2f}".format(r2_test))
```

## Bagging

```python
from sklearn.model_selection import GridSearchCV, cross_val_score, train_test_split
from sklearn.ensemble import BaggingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
from sklearn.model_selection import RandomizedSearchCV
import warnings
warnings.simplefilter(action="ignore", category=FutureWarning)

X_train, X_test, y_train, y_test = train_test_split(X, y.values, test_size=0.2, random_state=42)
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_samples': [0.5, 0.7, 0.9],
    'max_features': [0.5, 0.7, 0.9],
    'bootstrap': [False,True],
}

base_estimator = DecisionTreeRegressor(max_depth=4)

regressor_bg = BaggingRegressor(base_estimator=base_estimator, random_state=0)


gs = GridSearchCV(regressor_bg, param_grid, scoring='neg_mean_squared_error')
gs.fit(X, y.values)

best_params = gs.best_params_

regressor_bg.fit(X, y.values)
y_pred_train = gs.predict(X)
rmse_train = np.sqrt(mean_squared_error(y, y_pred_train))

y_pred_test = gs.predict(X_test)
rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_test))

best_n_estimators = gs.best_estimator_.n_estimators

r2_train = r2_score(y, y_pred_train)

r2_test = r2_score(y_test, y_pred_test)

print("Best Hyperparameters:", best_params)
print("Root Mean Squared Error (Training): {:.02f}".format(rmse_train))
print("Root Mean Squared Error (Testing): {:.02f}".format(rmse_test))
print("Best Number of Estimators:", best_n_estimators)
print("R^2 (Training): {:.02f}".format(r2_train))
print("R^2 (Testing): {:.02f}".format(r2_test))
```
